```python
df.drop(['Unnamed: 32','id'], axis = 1 , inplace=True)
df.diagnosis.replace({"M":1,"B":0},inplace=True)
```

```python
corr = df.corr()
fig, ax = plt.subplots(figsize=(20,20))
sns.cubehelix_palette(as_cmap=True)
sns.heatmap(df.corr().abs(),  annot=True, cmap = "BuPu", center = 0 )
```

```python
colors = ["#F19BF2","#A988F2" ]
sns.set_palette(sns.color_palette(colors))
f = sns.countplot(x='diagnosis', data=df)
f.set_title("Breast cancer")
f.set_xticklabels(['Benign', 'Malignant'])
plt.xlabel("");
```

```python
df_imp = df[['diagnosis', 'radius_mean', 'perimeter_mean', 'area_mean',
       'concavity_mean', 'concave points_mean', 'radius_worst',
       'perimeter_worst', 'area_worst', 'concave points_worst']]
```

```python
sns.pairplot(data=df_imp, hue="diagnosis", corner=True).fig.suptitle('Pairplot for Featrues', fontsize = 20)
```

```python
df.drop('diagnosis', axis=1).corrwith(df.diagnosis).plot(kind='bar', grid=True, figsize=(20, 10), title="Correlation with target",color="#A988F2");
```

```python
dfM = df[df['diagnosis'] == 1]
dfB = df[df['diagnosis'] == 0]
cols = df_imp.drop("diagnosis",axis=1).columns
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 10))
axes = axes.ravel()
for i, ax in enumerate(axes):
    ax.figure
    binwidth = (max(df[cols[i]]) - min(df[cols[i]])) / 50
    ax.hist([dfM[cols[i]], dfB[cols[i]]],
            bins=np.arange(min(df[cols[i]]), max(df[cols[i]]) + binwidth, binwidth), alpha=0.5,
            stacked=True, label=['M', 'B'], color=["#F19BF2", "#A988F2"])
    ax.legend(loc='upper right')
    ax.set_title(cols[i])
plt.tight_layout()
plt.show()  
```
